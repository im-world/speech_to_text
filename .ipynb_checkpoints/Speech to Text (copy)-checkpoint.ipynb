```python
import os
from pydub import AudioSegment
import whisper
from pyannote.audio import Pipeline
from tqdm import tqdm  # For progress bars

# Combine audio files into one MP3
def combine_audio_files(input_files, output_file):
    combined = AudioSegment.empty()
    for file in tqdm(input_files, desc="Combining audio files", unit="file"):
        audio = AudioSegment.from_file(file, format="aac")
        combined += audio
    combined.export(output_file, format="mp3")
    print(f"Combined audio saved as {output_file}")

# Diarization: Detect speakers in the audio
def diarize_audio(diarization_pipeline, audio_path):
    diarization = diarization_pipeline(audio_path)
    speaker_segments = []
    for segment in tqdm(diarization.itertracks(yield_label=True), desc="Processing speaker diarization", unit="segment"):
        speaker_segments.append((segment[0].start, segment[0].end, segment[1]))
    return speaker_segments

# Merge diarization with Whisper transcription
def merge_transcription_with_speakers(transcription, speaker_segments):
    labeled_text = ""
    for segment in tqdm(transcription["segments"], desc="Merging transcription and speakers", unit="segment"):
        start = segment["start"]
        end = segment["end"]
        text = segment["text"]

        # Find corresponding speaker label
        speaker = "Unknown"
        for speaker_start, speaker_end, speaker_label in speaker_segments:
            if start >= speaker_start and end <= speaker_end:
                speaker = "Viney: " if speaker_label == "FEMALE" else "Sandeep: "  # Map to A/B
                break

        labeled_text += f"{speaker}: {text}\n"
    return labeled_text

# Paths and filenames
input_files = ["WhatsApp Audio 2025-01-07 at 13.07.14.aac", 
               "WhatsApp Audio 2025-01-07 at 13.08.29.aac",
               "WhatsApp Audio 2025-01-07 at 13.14.53.aac"]
combined_output = "combined_audio.mp3"

# Combine the audio files
combine_audio_files(input_files, combined_output)

# Initialize Whisper and Pyannote models
print("Loading models...")
whisper_model = whisper.load_model("large", device='cpu')
# diarization_pipeline = Pipeline.from_pretrained("pyannote/speaker-diarization", use_auth_token="")

# Run Whisper transcription
print("Transcribing audio with Whisper...")
whisper_result = whisper_model.transcribe(combined_output)

with open('dr_bala_interview.txt', "w", encoding="utf-8") as f:
    f.write(whisper_result['text'])

# # Perform speaker diarization
# print("Performing speaker diarization...")
# speaker_segments = diarize_audio(diarization_pipeline, combined_output)

# # Combine transcription with speaker labels
# print("Merging transcription with speaker labels...")
# final_transcription = merge_transcription_with_speakers(whisper_result, speaker_segments)

# # Save the labeled transcription
# output_text_file = "transcription_with_speakers.txt"
# with open(output_text_file, "w") as f:
#     f.write(final_transcription)

print(f"Final transcription saved as {output_text_file}")
```

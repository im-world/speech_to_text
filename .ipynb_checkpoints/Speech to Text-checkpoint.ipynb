```python
import subprocess

def concatenate_mp4_to_mp3(file_list, output_file):
    # Create a text file with the list of input files
    with open("file_list.txt", "w") as f:
        for file in file_list:
            f.write(f"file '{file}'\n")
    
    # Run the ffmpeg command to extract and concatenate audio
    command = [
        "ffmpeg", "-f", "concat", "-safe", "0", "-i", "file_list.txt",
        "-vn", "-acodec", "libmp3lame", "-b:a", "192k", output_file
    ]
    
    try:
        subprocess.run(command, check=True)
        print(f"Audio extracted and concatenated into {output_file}")
    except subprocess.CalledProcessError as e:
        print(f"Error: {e}")
    finally:
        # Clean up the temporary file list
        import os
        if os.path.exists("file_list.txt"):
            os.remove("file_list.txt")

# Example usage
mp4_files = ["WhatsApp Audio 2024-11-29 at 17.47.28.mp4", "WhatsApp Audio 2024-11-29 at 17.47.29.mp4", 
               "WhatsApp Audio 2024-11-29 at 17.47.29(1).mp4", "WhatsApp Audio 2024-11-29 at 17.47.29(2).mp4", 
               "WhatsApp Audio 2024-11-29 at 17.47.30.mp4", "WhatsApp Audio 2024-11-29 at 17.47.30(1).mp4", 
               "WhatsApp Audio 2024-11-29 at 17.47.30(2).mp4", "WhatsApp Audio 2024-11-29 at 17.47.31.mp4"]
mp4_files = []
output_mp3 = "output.mp3"
concatenate_mp4_to_mp3(mp4_files, output_mp3)
```
